```python
def secondmax(arr):
    sublist = [x for x in arr if x < max(arr)]
    return max(sublist)

def secondmin(arr):
    listing = [x for x in arr if x > min(arr)]
    return min(listing)

if __name__ == '__main__':
    arr = [10, 20, 4, 45, 99]
    print("this is second max ",secondmax(arr))
    print("this is second min ",secondmin(arr))
```
